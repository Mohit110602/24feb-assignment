```python
#11)
date1=input('Enter the first date in YYYY-MM-DD format: ')
date1=pd.to_datetime(date1)
date2=input('Enter the second date in YYYY-MM-DD format: ')
date2=pd.to_datetime(date2)
dt=pd.date_range(start=date1,end=date2)
df1=pd.DataFrame(dt,columns=['Date'])
df1['Weekday_no.']=df1['Date'].dt.dayofweek
def weekdays(x):
    if x==0:
        return 'Monday'
    elif x==1:
        return 'Tuesday'
    elif x==2:
        return 'Wednesday'
    elif x==3:
        return 'Thursday'
    elif x==4:
        return 'Friday'
    elif x==5:
        return 'Saturday'
    elif x==6:
        return 'Sunday'
    else:
        return 'Error'
df1['Weekday']=df1['Weekday_no.'].apply(weekdays)
df1.drop('Weekday_no.',axis=1,inplace=True)
df1
```
